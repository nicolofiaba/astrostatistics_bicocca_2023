```python
import numpy as np
from numpy import random
import matplotlib.pyplot as plt
import math

#---------------------------------------------------
if "setup_text_plots" not in globals():
    from astroML.plotting import setup_text_plots
setup_text_plots(fontsize=14, usetex=True)
%config InlineBackend.figure_format='retina'
```

```python
N = 10000
x_min = 0.1
x_max = 10

x = random.uniform(x_min, x_max, N)
plt.hist(x, bins=50, density=True); # "Density=True" accounts for the normalization of the histogram
```

```python
y = np.log10(x)
plt.hist(y, bins=50, density=True);
```

We want to find the probability distribution for $y = log_{10}{x}$. Since $x$ comes from a uniform distribution, $p(x) = \frac{1}{b-a}$. 

We can apply the formula for *probability distribution transformation* to recover $p(y)$, obtaining: 
$$p(y) = \left| \frac{dx}{dy}\right| p(x) = 10^y \cdot log{10} \cdot \frac{1}{b-a}$$ 

```python
def pdf_y(y, a, b):
        return (10**(y) * np.log(10) * (1/(b-a)))

x_array = np.linspace(0.1, 10, 100)
y_array = np.log10(x_array)
pdf_y = pdf_y(y_array, 0.1, 10)
plt.hist(y, bins=50, density=True);
plt.plot(y_array, pdf_y)
```

```python
print("The logarithm of the mean of the samples is: ", np.log10(np.mean(x)))
print("The mean of the logs of the samples is: ", np.mean(y))

print("\nThe logarithm of the median of the samples is: ", np.log10(np.median(x)))
print("The median of the logs of the samples is: ", np.median(y))
```
