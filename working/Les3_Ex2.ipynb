```python
import numpy as np
from pylab import plt
from scipy.stats import poisson
```

We expect the discrete variable "*number of deaths*" to be distributed according to a Poisson distribution,
$$ p(k|\mu) = \frac{\mu^k \exp(-\mu)}{k!} $$

with $\mu$ being the unknown true *death probability*, which we can estimate by computing a weighted average of the number of deaths in our dataset.

```python
total = 200
number_of_deaths = np.arange(0, 5, dtype=int)
number_of_groups = np.array([109, 65, 22, 3, 1])
frequencies = number_of_groups/total

plt.scatter(number_of_deaths, frequencies, label="Death frequencies");

# We estimate now estimate the death proability
death_prob = np.average(number_of_deaths, weights=number_of_groups)

Poisson = poisson(death_prob)
x = np.arange(0, 5, dtype=int)
plt.scatter(x, Poisson.pmf(x), label=r"Poisson distribution values ($\mu_{est} = %.2f$)"%death_prob);
plt.grid(True, ls='--', alpha=0.5)
plt.legend();
```

We can clearly notice that the data are well fitted by a Poissonian distribution, as expected.
